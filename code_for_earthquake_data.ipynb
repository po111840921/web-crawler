```python
#幫檔案改名
#要先將下載名稱依時間排序

import os

def batch_rename(path):
    year = 2012
    month = 1
    for fname in os.listdir(path):
        new_fname = f"{year}_{month}.csv"
        os.rename(os.path.join(path, fname), os.path.join(path, new_fname))
        if month == 12:
            year += 1
            month = 1
        else:
            month += 1

batch_rename("C:\\python\\project2\\")
```
